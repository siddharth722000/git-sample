# operators
## 1.arithmetic operators ----->+,-,*,/,//,%,**
## 2.logical operators ------>and ,or ,not
## 3.relational operators ---------> <,>,<=,>=,==,!=
## 4.membership operators -------> in,not in


```python
#aritmentic operators
a=10
b=20
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(a**b)
```

```python
#relational operators 
x=10
y=20
print(x==y)
print(x!=y)
print(x<y)
print(x>y)
```

```python
#logical operators 
x=20
y=30
print(x>=10 and x<=20)
print(y>=5 or y<=20)
print(not x)
```

```python
#membership operators
str1 ='abc'
print('a'in  str1)
print('d'not in str1)
```

```python
# to check whether the given number is positive ,negative and zero
n=int(input('enter a number'))
if n==0:
    print('zero')
elif n>0:
    print('positive number')
elif n<0:
    print('negative number')
```
